```python
import pandas as pd
from numpy import median
import numpy as np
```

```python

CSK=pd.read_csv(f"./Graphs/CSK/CSK.csv").T
KKR=pd.read_csv(f"./Graphs/KKR/KKR.csv").T
DC=pd.read_csv(f"./Graphs/DC/DC.csv").T
RCB=pd.read_csv(f"./Graphs/RCB/RCB.csv").T
df=pd.concat([CSK,KKR,DC,RCB],axis=1)
df.columns=["CSK","KKR","DC","RCB"]
df.drop(df.index[0],inplace=True)
# df.index[0]
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
rows = 1
columns = 4
fig = plt.figure(figsize=(30, 9))
CSK = cv.imread('./Graphs/CSK/CSK_Directed.png')
DC = cv.imread('./Graphs/DC/DC_Directed.png')
RCB = cv.imread('./Graphs/RCB/RCB_Directed.png')
KKR = cv.imread('./Graphs/KKR/KKR_Directed.png')

fig.add_subplot(rows, columns, 1)
plt.axis('off')
plt.title("CSK_Directed")
CSK=cv.cvtColor(CSK, cv.COLOR_BGR2RGB)
plt.imshow(CSK)


fig.add_subplot(rows, columns, 2)
plt.axis('off')
plt.title("DC_Directed")
DC=cv.cvtColor(DC, cv.COLOR_BGR2RGB)
plt.imshow(DC)


fig.add_subplot(rows, columns, 3)
plt.axis('off')
plt.title("KKR_Directed")
KKR=cv.cvtColor(KKR, cv.COLOR_BGR2RGB)
plt.imshow(KKR)


fig.add_subplot(rows, columns, 4)
plt.axis('off')
RCB=cv.cvtColor(RCB, cv.COLOR_BGR2RGB)
plt.title("RCB_Directed")
plt.imshow(RCB)  

plt.show()
```
